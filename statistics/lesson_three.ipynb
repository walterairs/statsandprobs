3.1 Laske tehtävän 1.3 tilastomuuttujalle vaihteluvälin pituus, keskipoikkeama, keskihajonta, varianssi ja otoskeskihajonta. Huom, voit käyttää aiemmin laskemaasi keskiarvoa.


```python
## Code here
###Code here
weight_frequency_kg = {5:12, 10:17, 15:13, 20:8, 25:17, 30:5}

## Create a list of weights, so the calculation of average, median, and mode is easier
list_of_weights = []
for weight, frequency in weight_frequency_kg.items():
    for i in range(frequency):
        list_of_weights.append(weight)

## Calculate average, median, and mode
average = numpy.mean(list_of_weights)
median = numpy.median(list_of_weights)
mode = pandas.Series(list_of_weights).mode()[0]

data = pandas.DataFrame({'Weight (kg)': list_of_weights})

## calculate the frequency of each weight
frequency = data['Weight (kg)'].value_counts()
data2 = pandas.DataFrame({'Weight (kg)': frequency.index, 'Frequency': frequency.values})

# lets calculate mean deviation, variance, lenght of range, standard deviation and
# sample standard deviation
mean_deviation = numpy.mean(numpy.absolute(data2['Weight (kg)'] - average))
variance = numpy.var(data2['Weight (kg)'])
d_range = numpy.max(data2['Weight (kg)']) - numpy.min(data2['Weight (kg)'])
standard_deviation = numpy.sqrt(variance)
sample_standard_deviation = numpy.sqrt(variance * (len(list_of_weights) - 1) / len(list_of_weights))

# lets plot
plt.figure(figsize=(12, 8))
plt.hist(list_of_weights, bins=6, edgecolor='black')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.title('Frequency of Weights')
plt.show()

print('Average: ', average)
print('Median: ', median)
print('Mode: ', mode)
print('Mean Deviation: ', mean_deviation)
print('Variance: ', variance)
print('Range: ', d_range)
print('Standard Deviation: ', standard_deviation)
print('Sample Standard Deviation: ', sample_standard_deviation)
```

3.2 Laske tehtävän 1.4 tilastomuuttujalle keskihajonta ja otoskeskihajonta.


```python
## Code here
cat_age_frequency = {16:1, 17:2, 18:3, 19:2, 20:1}

## Create a list of cat ages, so the calculation of average, median, and mode is easier
list_of_cat_ages = []
for cat_age, frequency in cat_age_frequency.items():
    for i in range(frequency):
        list_of_cat_ages.append(cat_age)

average = numpy.mean(list_of_cat_ages)

print(f"average: {average}")

# lets calculate standard deviation and sample standard deviation
variance = numpy.var(list_of_cat_ages)
standard_deviation = numpy.sqrt(variance)

print(f"variance: {variance}")

sample_standard_deviation = numpy.sqrt(variance * (len(list_of_cat_ages) - 1) / len(list_of_cat_ages))

print(f"sample standard deviation: {sample_standard_deviation}")

# lets plot
plt.figure(figsize=(12, 8))
plt.hist(list_of_cat_ages, bins=5, edgecolor='black')
plt.xlabel('Cat Age')
plt.ylabel('Frequency')
plt.title('Frequency of Cat Ages')
plt.show()
```
