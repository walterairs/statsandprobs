8.3 Alla olevan aineiston pit√§isi noudattaa regressiosuoraa ùë¶=3ùë•+5,7. Havaintoarvo (5,28) on kuitenkin poikkeavan virheellinen, mik√§ on sen absoluuttinen ja suhteellinen virhe?


```python
## Code here
x = [x for x in range(1, 8)]
y = [8.7, 11.7, 14.7, 17.7, 28, 23.7, 26.7]

plt.plot(x, y, 'ro')

def regline(x):
    return 3 * x + 5.7

# just show the line
plt.plot(x, [regline(x) for x in x], 'b-')

absolute_error = 28 - regline(5)
relative_error = (absolute_error / 28) * 100

print("Absolute error: ", absolute_error)
print("Relative error: ", relative_error, "%")
```

8.4 Mik√§ on edellisen teht√§v√§n satunnaisen virheen suuruus? Jos edellisen teht√§v√§n aineiston pit√§isi noudattaa regressiosuoraa ùë¶=3ùë•+2, kuinka suuri olisi systemaattinen virhe?

```python
## Code here
deg_of_explanation = (numpy.corrcoef(x, y)[0, 1]) ** 2
print("Degree of explanation: ", deg_of_explanation * 100, "%")
print("random_error: ", (1 - deg_of_explanation) * 100, "%")

def regline2(x):
    return 3 * x + 2

# just show the line
plt.plot(x, y, 'ro')
plt.plot(x, [regline2(x) for x in x], 'r-')
plt.plot(x, [regline(x) for x in x], 'g-')
plt.show()

systematic_error = 3.7

print("Systematic error: ", systematic_error)
```
